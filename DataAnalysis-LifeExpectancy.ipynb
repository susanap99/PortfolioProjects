## Life Expectancy Evolution

```python
df_port = df.loc[df['Country']=='Portugal']

sns.lineplot(x= 'Year', y= 'Life expectancy', data=df, color='r', label='World')
sns.lineplot(x = 'Year', y = 'Life expectancy', data=df_port, color='orange', lw=2, label='Portugal')

plt.title('Life Expectancy Evolution')

plt.xlabel('Year')
plt.ylabel('Life Expectancy (age)')

plt.show()
```

##### In general, life expectancy in the world has been increasing since 2000. In Portugal, the life expectancy has always been above average, and also been increasing over the years. There was, however, a decrease from 2014 to 2015.

### Question 1: What variables affect the life expectancy?

###### Socio-economic variables to analyse:
percentage expenditure, Total expenditure, GDP, Income composition of resources, Schooling

###### Demographic variables to analyse:
Adult Mortality, infant deaths, under-five deaths, Population

##### Health variables to analyse:
Alcohol, Hepatitis B, Measles, BMI, Polio, Diphteria, HIV/AIDS, thinness 1-19 years, thinness 5-9 years 

```python
#Socio-economic variables

df_e = df[['Life expectancy','percentage expenditure', 'Total expenditure', 'GDP', 'Income composition of resources', 'Schooling']]
df_e.corr()
```

```python
e_matrix = df_e.corr().round(2)
sns.heatmap(e_matrix, annot=True, cmap ='OrRd')
plt.show()
```

##### Here, we see that both Schooling and Income Composition of Resources are the socio-economic variables that affect   Life Expectancy the most. The more schooling and income composition of resources, the longer is the life expectancy in a country.

```python
# Demographic variables

df_d = df[['Life expectancy', 'Adult Mortality', 'infant deaths', 'under-five deaths', 'Population']]

df_d.corr()
```

```python
d_matrix = df_d.corr().round(2)
sns.heatmap(d_matrix, annot=True, cmap='OrRd')
plt.show()
```

##### As for demographic variables, we conclude that Adult Mortality affects Life Expectancy the most. As the adult mortality decreases, the life expectancy increases.

##### Infant mortality has a negative correlation to life expectancy as well, but not as much as adult mortality.

```python
# Health variables

df_h = df[['Life expectancy', 'Alcohol', 'Hepatitis B', 'Measles', 'BMI', 'Polio',
           'Diphteria', 'HIV/AIDS', 'thinness 1-19 years', 'thinness 5-9 years']]

df_h.corr()
```

```python
h_matrix = df_h.corr().round(2)
sns.heatmap(h_matrix, annot=True, cmap='OrRd')
plt.show()
```

##### Regarding health variables, we see that BMI and HIV/AIDS are the most correlated to life expectancy. A low BMI means less life expectancy; the less people infected with HIV/AIDS, the more is life expectancy in a country.

##### Alcohol has a positive correlation with life expectancy. Although it was expected for Alcohol to have a negative correlation with Life Expectancy, that can be explained by the fact that low-income countries will not have much resources to afford alcohol.

##### Diphteria and Polio also have a positive correlation with Life Expectancy. For both, there are vaccines that help prevent death. The same doesn't happen to HIV, wich has no preventive vaccine.

### Question 2: Should a country having a lower life expectancy value (<65)  increase its healthcare expenditure in order to improve its average lifespan?


```python
df_avg = df.groupby(['Country']).mean()

df_avg
```

```python
plt.scatter(x=df_avg['Life expectancy'], y=df_avg['percentage expenditure'], color='orange')

plt.title('Life Expectancy vs Healthcare Expenditure')

plt.xlabel('Life Expectancy (age)')

plt.ylabel('Healthcare Expenditure (as % of GDP)')

plt.show()
```

###### In average for each country, life expectancy increases as healthcare expenditure increases. However, we only see a greater healthcare expenditure for countries with life expectancy over 75 years. Countries with an average life expectancy of <65 years, could increase its healthcare expenditure in order to improve lifespan, although as we've seen before, there are other economic variables that influence life expectancy more (such as schooling or income composition of resources).
